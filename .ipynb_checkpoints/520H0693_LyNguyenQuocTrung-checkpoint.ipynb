```python
import nltk
import re
from nltk import ngrams
from collections import Counter
```

```python
nltk.download('punkt')
```

```python
filename = 'train.txt'
with open(filename, encoding = 'utf-8') as f:
  lines = f.readlines()
print(lines)
```

```python
counter_bigram = Counter()
for sent in lines:
  tokens = nltk.word_tokenize(sent.lower())
  tokens = ['<s>'] + tokens
  gram2 = ngrams(tokens,2)
  join_grams = ['_'.join(list(gram)) for gram in gram2]
  counter_bigram.update(join_grams)
print(len(counter_bigram))
print(counter_bigram.most_common(10))
```

```python
counter_unigram = Counter()
for sent in lines:
  tokens = nltk.word_tokenize(sent.lower())
  tokens = ['<s>'] + tokens
  gram2 = ngrams(tokens,1)
  join_grams = ['_'.join(list(gram)) for gram in gram2]
  counter_unigram.update(join_grams)
print(len(counter_unigram))

V = len(counter_unigram)
print(counter_unigram.most_common(10))

total = 0
for i in counter_unigram.values():
  total +=i
print(total)
```

```python
def Psent(sent):
  prob = 1
  alpha =1.01
  tokens = nltk.word_tokenize(sent.lower())
  tokens = ['<s>'] +tokens
  for i in range(1,len(tokens)):
    prob *= ((counter_bigram[tokens[i-1]+'_'+tokens[i]])+alpha)/((counter_unigram[tokens[i-1]])+V*alpha)
  return prob
```

```python
def uniPsent(sent):
  prob = 1
  alpha =1.01
  tokens = nltk.word_tokenize(sent.lower())
  tokens = ['<s>'] +tokens
  for i in range(len(tokens)):
    prob *= ((counter_bigram[tokens[i]])+alpha)/(total+(V*alpha))
  return prob
```

```python
text = "i like a book"
text2 = "i like a famous book"
print(Psent(text))
print(Psent(text2))
print(uniPsent(text))
```
